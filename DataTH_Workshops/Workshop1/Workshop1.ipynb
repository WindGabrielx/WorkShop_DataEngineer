```python
#@title ตั้งค่า Config เข้าสู่ข้อมูล DataBase
import os
class Config:
  MYSQL_HOST =
  MYSQL_PORT =
  MYSQL_USER =
  MYSQL_PASSWORD =
  MYSQL_DB =
  MYSQL_CHARSET =


print(Config.MYSQL_PORT) #เช็คว่าเชื่อมต่อได้หรือไม่
```

```python
#@title ทำการเข้าสู่ข้อมูลโดยอ้างอิงเงื่อนไขก่อนหน้าโดยใช้ Pymysql
import pymysql
connection  = pymysql.connect(host=Config.MYSQL_HOST,
                             port=Config.MYSQL_PORT,
                             user=Config.MYSQL_USER,
                             password=Config.MYSQL_PASSWORD,
                             db=Config.MYSQL_DB,
                             charset=Config.MYSQL_CHARSET,
                             cursorclass=pymysql.cursors.DictCursor)
```

```python
#@title เทสว่า Connect รึยัง
connection
```

```python
#@title ทำการเช็คว่าใน Tables มีอะไรบ้างซึ่งที่นี้จะแบ่งเป็น  2  ประเภท
# list all tables ด้วย SQL คำสั่ง show tables;
cursor = connection.cursor()
cursor.execute("show tables;")
tables = cursor.fetchall()
cursor.close()
print(tables)
```

```python
#@title Query Table
with connection.cursor() as cursor:
  cursor.execute("SELECT * FROM audible_data;")
  result = cursor.fetchall() #เก็บค่า result ในรูปแบบตัวแปร result

#print(result)#เพื่อเช็คว่าได้ไหม
print("number of rows:",len(result)) #มีจำนวน Rows ทั้งหมด 2269
```

```python
with connection.cursor() as cursor:
  cursor.execute("SELECT * FROM audible_transaction;")
  result2 = cursor.fetchall() #เก็บค่า result ในรูปแบบตัวแปร result

#print(result)#เพื่อเช็คว่าได้ไหม
print("number of rows:",len(result2)) #มีจำนวน Rows ทั้งหมด 1998821
```

```python
#@title Convert to Pandas
import pandas as pd
audible_data = pd.DataFrame(result)
audible_data
```

```python
audible_transaction = pd.DataFrame(result2)
audible_transaction
```

```python
#@title กำหนดให้ Book_ID เป็น index เพื่อทำการ merge กัน
audible_data = audible_data.set_index("Book_ID")
```

```python
audible_data
```

```python
#@title นำเอาข้อมูลทั้งสองมา Merge กันโดยใช้ Index Book_ID

transaction = audible_transaction.merge(audible_data, how="left", left_on="book_id", right_on="Book_ID")
transaction
```

```python
#@title ทำการสร้างตัวแปรเพิ่ม โดยการ copy timestamp และสร้างคอลัมน์ชื่อ date เพิ่ม
transaction['date'] = transaction['timestamp']
transaction
```

```python
#@title เนื่องจากค่าเงินยังไม่มีการแปลงเลยทำการ Import ตัวแปลงเงินมาโดยใช้คำสั่ง Requests
# ทำการ import ตัวแปลงเงิน
import requests
url = "https://r2de2-workshop-vmftiryt6q-ts.a.run.app/usd_thb_conversion_rate"
r = requests.get(url)
result_conversion_rate = r.json()
```

```python
print(type(result_conversion_rate))
assert isinstance(result_conversion_rate, dict) #เป็นการเช็คคำสั่งหาก result_conversion_rate เป็น dict จะเป็น True
```

```python
conversion_rate = pd.DataFrame(result_conversion_rate)
conversion_rate
```

```python
#@title  ทำการกำหนด index เป็น date
conversion_rate = conversion_rate.reset_index().rename(columns={"index": "date"})
conversion_rate[:3]
```

```python
transaction
```

```python
# แปลงให้จาก timestamp เป็น date ในทั้ง 2 dataframe (transaction, conversion_rate) เนื่องจากบางหัวข้อ date ยังเป็น str
transaction['date'] = pd.to_datetime(transaction['date']).dt.date
conversion_rate['date'] = pd.to_datetime(conversion_rate['date']).dt.date
transaction.head()
```

```python
#@title นำตาราง transaction, conversion_rate เข้ารวมด้วยกันและเชื่อมกันด้วย date
#กำหนดตัวแปรสุดท้ายคือ final_df
final_df = transaction.merge(conversion_rate,how = "left", left_on="date", right_on="date")
final_df
```

```python
#ต้องการนำ $ ออก และแปลง Price เป็นข้อมูลแบบ float
final_df["Price"] = final_df.apply(lambda x: x["Price"].replace("$",""), axis=1)
final_df["Price"] = final_df["Price"].astype(float)
final_df
```

```python
#@title ทำการสร้างตาราง THBPrice โดยเกิดจากการนำ Column Price * Conversion_rate
final_df["THBPrice"] = final_df["Price"] * final_df["conversion_rate"]
final_df
```

```python
final_df.to_csv("windtest1.csv", index = False)
```

```python
top_categories = final_df.groupby("Categories")['THBPrice'].sum().nlargest(3)
print(top_categories)
```

```python
filtered_df = final_df[final_df["Categories"] == "Literature & Fiction"]
test = filtered_df.groupby("Book Title")['THBPrice'].sum().nlargest(3)
print(test)
#เนืองจากเรารู้ว่า Categories ชื่อ Literature & Fiction มีจำนวนเงินมากสุด
```
